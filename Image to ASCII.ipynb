## Today we will use PyWhatkit library and convert a image to ASCII Code.

```python
import pywhatkit as kt
```

#### First install PyWhatkit library and then import the same.

```python
source_path='pokemon_PNG148.png'
destination_path='asciiart.txt'
```

#### Specify the image file name i.e the source path and the destination path as of where to store the ascii code.

```python
kt.image_to_ascii_art(source_path,destination_path)
```

### A txt file would have been created with the name as specified in the destination.
